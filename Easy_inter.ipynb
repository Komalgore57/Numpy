Create a checkerboard pattern matrix.

```python
checkboard = np.zeros((4,4))
checkboard[::2,1::2]=1
checkboard[1::2,::2]=1
checkboard
```

Create a diagonal matrix from an array.

```python
checkboard = np.zeros((4,4))
for i in range(0,4):
    checkboard[i,i]=1

checkboard
```

Extract the diagonal elements of a matrix.

```python
for i in range(0,4):
    print(checkboard[i,i])
```
